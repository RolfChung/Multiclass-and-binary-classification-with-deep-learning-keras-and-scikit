## Binary classification modeling with Deep Learning neural networks

<img src="neuralnetworkweights.png" alt="Smiley face" height="500" width="500"> 
<p>A deep learning neural network consists of an input layer, at least one hidden
layer and one output layers. The neurons or links connected the nodes of the
subsequent layers. Neurons carry weights (parameters). At each node the activation
function takes the inputs carried forward by multiple neurons and
calculates an output, which is turn carried forward to the nodes of the next layer.
This is done till the output layer is reached and forward propagation is
completed. </p> 

<p>The process of backward propagation optimizes the process 
using the gradient descent method. This method manipulates the weights (gradients) of the
network to make better output predictions with a lower error, cost or loss than before.
The error at the bottom of the gradient error descent is the optimal or minimal error.
Particular loss functions are better suited for particular problems. It is
a topic of research to develop algorithm with higher performance for special
tasks.</p> 


<img src="gradientdescent.png" alt="Smiley face" height="500" width="500"> 

<p>A question of particular interest are the learning steps taken by the 
algorithm to find the minimum. Greater learning steps will yield 
quicker results, but can miss or jump over the minimum. 
It is helpful to remember that the idea of gradient descent
relies on linear algebra and differentiation like this:</p>

<img src="Costfunction.jpg" alt="Geeks for geeks" height="400" width="300"> 

<p>
<a href="https://www.geeksforgeeks.org/gradient-descent-in-linear-regression/">
Geeks for geeks
</a> 
</p>

<p>
The learning rate alpha alpha is a multiplicator that produces the size of the learning
step, when sliding down the downward slope to the local minima by adjusting the weights.
Higher learning rates result in higher step down the slope.
The risk is then to miss local minima.
This is bad because local minima are optimal values.
Missing this means to take values on the upward slope side into account.
</p>

#### Making a function for plotting the dashboard

```python
def plot_dashboard(X_values, y_values):
        
        fig, ax = plt.subplots(2,2,figsize=(15,10))
        
        plt.suptitle('Dashboard ')

        ax[0,0].plot(X_values, y_values.history['val_loss'], color='olivedrab', linewidth=3)
        ax[0,0].plot(X_values, y_values.history['loss'], color='crimson', linewidth=3)
        ax[0,0].set_title('loss')
        ax[0,0].set_xlabel('epoch')
        ax[0,0].legend(['test', 'train'])
        #ax[0,0].set_xticks(X_values)

        ax[0,1].plot(X_values, y_values.history['val_accuracy'], color='olivedrab', linewidth=3)
        ax[0,1].plot(X_values, y_values.history['accuracy'], color='crimson', linewidth=3)
        ax[0,1].set_title('accuracy')
        ax[0,1].set_xlabel('epoch')
        ax[0,1].legend(['test', 'train'])
        #ax[0,1].set_xticks(X_values)

        ax[1,0].plot(X_values, y_values.history['loss'], color='olivedrab', linewidth=3)
        ax[1,0].plot(X_values, y_values.history['accuracy'], color='crimson', linewidth=3)
        ax[1,0].set_title('accuracy and loss of training data')
        ax[1,0].set_xlabel('epoch')
        ax[1,0].legend(['loss', 'accuracy'])
        #ax[1,0].set_xticks(X_values)


        ax[1,1].plot(X_values, y_values.history['val_loss'], color='olivedrab', linewidth=3)
        ax[1,1].plot(X_values, y_values.history['val_accuracy'], color='crimson', linewidth=3)
        ax[1,1].set_title('accuracy and loss of validation data')
        ax[1,1].set_xlabel('epoch')
        ax[1,1].legend(['validation loss', 'validation accuracy'])
        # ax[1,1].set_xticks(X_values)


        return plt.show()
```

## Multiclass classification

<p>

Multiclass classification makes the assumption that each sample is assigned to 
one and only one label: a fruit can be either an apple or a pear but 
not both at the same time." The necessary condition for "Multiclass classification" is that 
all labels are mutually exclusive and unambiguous.
</p>

on all four quadrants. It depends on the probability threshold above 0.5 on which areas the attack is carried out.

```python
def generic_model(acttype):
    
    generic_model = Sequential()
    # the input consists of 2 variables
    generic_model.add(Dense(120, input_shape=(2,), activation=acttype))
    generic_model.add(Dense(60, activation=acttype))
    generic_model.add(Dense(30, activation=acttype))
    generic_model.add(Dense(15, activation=acttype))
    # This is the output layer with four classes
    generic_model.add(Dense(4, activation='softmax'))
    # 'categorical_crossentropy'
    generic_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return generic_model

# keras.layers.advanced_activations. These include PReLU and LeakyReLU
```

```python
activations = ['relu', 'sigmoid', 'tanh']

activations_performance = dict()

specificlosses = dict()
specificaccs = dict()

earlystopping_act = EarlyStopping(monitor='val_loss', patience=3)

for activation in activations:
    specific_model = generic_model(acttype=activation)
    specific_history = \
    specific_model.fit(darts_train_scaled, competitor_train, epochs=30, verbose=0,
                       validation_data=(darts_test_scaled, competitor_test))
    # optional: , callbacks=[earlystopping_act]
    # this produces dictionaries of different lengths
    
    
    activations_performance[activation] = specific_history
    
    specific_loss=specific_model.evaluate(darts_test_scaled, competitor_test, verbose=0)[0]
    specificlosses[activation] = specific_loss
    
    specific_acc=specific_model.evaluate(darts_test_scaled, competitor_test, verbose=0)[0]
    specificaccs[activation] = specific_acc
    
```

#### Does the batch size influence the model performance?

<p>
According to the doc, batch_size (Integer or None) is the number of samples per gradient update. 
If unspecified, batch_size will default to 32. Do not specify the batch_size if your data is in 
the form of symbolic tensors, generators, or Sequence instances (since they generate batches).
</p> 
<p>
A network with 5 batches as inputs, also called mini-batches, consisting of partition of observations
instead of the whole data set, is updated three times within one epoch. 
This can lead to a lower error rate as local loss minima  are updated 5 times, but 
leads to erratic swings, which can be countered with BatchNormalization.
Also naturally less RAM is required, because less observations are fed into the layers and
networks train and the network is training quicker.
</p> 
<p>
In this "experiment" the same model is fitted with different batch sizes.
</p>

### Does batch normalization increases the performance?

<p>
According to the doc, Batch normalization normalizes the activations of the previous layer at each batch, 
i.e. applies a transformation that maintains the mean activation close to 0 and the activation 
standard deviation close to 1.
</p> 
<p>
If you take a look at the pictures above it is clear that 0 is a crucial point for all of this function.
Centering it around 0 makes the outputs more stable by reducing the range of output values.
</p> 
<p>
Again two models are developed, one with and one without batch normalization, and compared.
</p> 

#### Developing the batchnormal_model

```python

def batchnormal_model(acttype):
    
        batchnormal_model = Sequential()
        # the input consists of 2 variables
        batchnormal_model.add(Dense(120, input_shape=(2,), activation=acttype, kernel_initializer='normal'))
        batchnormal_model.add(BatchNormalization())
        batchnormal_model.add(Dense(60, activation=acttype, kernel_initializer='normal'))
        batchnormal_model.add(BatchNormalization())
        batchnormal_model.add(Dense(30, activation=acttype, kernel_initializer='normal'))
        batchnormal_model.add(BatchNormalization())
        batchnormal_model.add(Dense(15, activation=acttype, kernel_initializer='normal'))
        
        # This is the output layer with four classes
        batchnormal_model.add(Dense(4, activation='softmax'))
        
        # 'categorical_crossentropy'
        batchnormal_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

        return batchnormal_model
```

#### Create a function with a keras model as input into the KerasClassifier-wrapper

```python
def hypertuned(learning_rate, activation='relu'):
    
    # the optimizers are offering an argument learning rate
    # learning_rate: float >= 0. Learning rate.
    # https://keras.io/optimizers/
    tuned_optimizer = keras.optimizers.adam(learning_rate=learning_rate)
    
    hypertuned_model = Sequential()
    hypertuned_model.add(Dense(60, input_shape=(2,), kernel_initializer='normal', activation=activation))
    hypertuned_model.add(BatchNormalization())
    hypertuned_model.add(Dense(120, kernel_initializer='normal', activation=activation))
    hypertuned_model.add(BatchNormalization())
    hypertuned_model.add(Dense(80, kernel_initializer='normal', activation=activation))
    hypertuned_model.add(BatchNormalization())
    hypertuned_model.add(Dense(40, kernel_initializer='normal', activation=activation))
    
    hypertuned_model.add(Dense(4, activation='softmax'))
    
    # using keras metrics
    # import keras_metrics as km
    # https://pypi.org/project/keras-metrics/
    
    hypertuned_model.compile(optimizer=tuned_optimizer,  loss='binary_crossentropy',
                             metrics=['accuracy', km.precision(), km.recall()])
    
    # hypertuned_random.fit(darts_train, competitor_train)
    
    return hypertuned_model

    
```

```python
# create the first model with the hypertuned function using a KerasClassifier as a wrapper
# build_fn: callable function or class instance
# KerasClassifier(build_fn = function)
hypertuned_wrapped_1 = KerasClassifier(build_fn = hypertuned)

# three parameters dictionary are build 
# the more parameters are set up, the longer the processing time
# keeping this in mind hyperparams3 with the lowest number of parameters in mantained as active code,
# while maybe not producing the best results

# hyperparams dictionary produces
# {'learning_rate': 0.1, 'epochs': 90, 'batch_size': 60, 'activation': 'relu'}
# 0.8176305838930074
# takes more than one hour under this conditions
hyperparams = {'learning_rate': [0.1, 0.01, 0.001],
               'batch_size' : [60, 120, 240],
               'epochs' : [30, 60, 90],
               'activation' : ['relu', 'tanh']}

# 0.797108223856385
# <keras.wrappers.scikit_learn.KerasClassifier object at 0x000002D027E5BFC8>
# {'learning_rate': 0.01, 'epochs': 90, 'batch_size': 60, 'activation': 'relu'}
hyperparams2 = {'learning_rate': [0.1, 0.01, 0.001],
               'batch_size' : [60, 120],
               'epochs' : [30, 90],
               'activation' : ['relu']}

# Results of hyperparams3
# 0.736473904283189
# <keras.wrappers.scikit_learn.KerasClassifier object at 0x000002CFA885B3C8>
# {'learning_rate': 0.1, 'epochs': 30, 'batch_size': 60, 'activation': 'relu'}
hyperparams3 = {'learning_rate': [0.1, 0.01, 0.001],
               'batch_size' : [60],
               'epochs' : [30],
               'activation' : ['relu']}

```
